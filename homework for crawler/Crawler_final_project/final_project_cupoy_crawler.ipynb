```python
# 打開瀏覽器
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.cupoy.com/newsfeed/topstory")

# 每個兩秒鐘自動往下滑
import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

# 設定list與迴圈
count = 1
title500 = []
category_list = [] # 先把類別放進來
category_count = {} # 整理各類別有多少個

while (count <= 500):
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    news = soup.find('div', class_ = 'ReactVirtualized__Grid__innerScrollContainer').find_all(class_='sc-eEieub sc-iuDHTM ibJqYc')
    
    for titles in news:
        category = titles.find(class_ = 'sc-gacfCG bPSpUf').text
        title = titles.find(class_ = 'sc-erNlkL sc-ekulBa hDLssh').text
        
        if title not in title500:
            print(count, category, title)
            count += 1
        title500.append({
        '類別': category,
        '標題': title
        })
        category_list.append(category)
        
        if count == 501:
            break
    
# 關閉瀏覽器
browser.quit();
```

```python
# 以圖表顯示
import pandas as pd
import numpy as np

pd.DataFrame(title500)
```

```python
# 總計類別，並選出最多類別
for category in category_list:
    if category not in category_count:
        category_count[category] = 1
    else:
        category_count[category] += 1

print('類別:', category_count)
print('類別最多為:', max(category_count, key = category_count.get))
```

```python
# way1類別視覺化
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

myfont = FontProperties(fname = r"./NotoSansCJK-Black.ttc")

# 顯示餅圖
# patches：得到餅圖的返回值
# l_texts：餅圖外標籤的文字
# p_texts：餅圖內部文字的value
patches,l_text,p_text = plt.pie(category_count.values(), labels=category_count.keys(), autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

for t in l_text: 
    t.set_fontproperties(myfont)
    
# plt.legend(loc = "right", prop=myfont)

plt.show()
```

```python
# way2類別視覺化
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

myfont = FontProperties(fname = r"./NotoSansCJK-Black.ttc")

# plt.figure(figsize=(10,15))
plt.plot('category_count')
plt.bar(range(len(category_count)), list(category_count.values()), align='center')
plt.xticks(range(len(category_count)), list(category_count.keys()), fontproperties = myfont)

plt.show()
```
