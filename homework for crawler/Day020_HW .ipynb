```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
table
```

```python
data = {}
key = ''
date = ''
items = table.find_all('td')
for item in items:
    if 'style' in item.attrs:
        key = item.text
        data[key]={}
    elif 'class' not in item.attrs:
        if item.text.strip() != '':
            if '/'in item.text:
                date = item.text
            else:
                data[key][date] = item.text
print(data)
```
